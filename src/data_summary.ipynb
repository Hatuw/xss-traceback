```python
import os
import pandas as pd
import matplotlib.pyplot as plt

DATA_DIR = '../data'
DATA_FILE = 'mirror.csv'
DATA_FILE = 'xss.csv'

data_file = os.path.join(DATA_DIR, DATA_FILE)
```

```python
data = pd.read_csv(data_file)
data.head()
```

```python
# pagerank distribution
# data['Pagerank'].hist(bins=30)
import numpy as np
log_pr = np.log(data[data['Pagerank'] != 0]['Pagerank'])
# plt.hist(data['Pagerank'], bins=30, log=True, )
plt.hist(log_pr, bins=30, log=True )
plt.xlabel("$log_e PR$")
plt.ylabel("Count")
plt.grid()
plt.show()
```

```python
from collections import Counter, OrderedDict
ct_result = dict(Counter(data['Author']))
ct_result
```

```python
# 画出作者提交payload数量-频数关系图
counter_values = list(ct_result.values())
plt.hist(counter_values, bins=50, log=True)
plt.show()
```
